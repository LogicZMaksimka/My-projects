```python
import sys
import matplotlib.pyplot as plt

# block_sizes_list = [2**2, 2**4, 2**6, 2**8, 2**10]
# array_length_list = [2**11, 2**14, 2**17, 2**20, 2**23, 2**25, 2**27]
# matrices_size_list = [2**5, 2**7, 2**9, 2**11, 2**13]

array_block_sizes_list = [2, 4, 6, 8, 10]
array_length_list = [11, 14, 17, 20, 23, 25, 27]
matrices_size_list = [5, 7, 9, 11, 13]
matrices_block_sizes_list = [1, 2, 3, 4, 5]
```

```python
def foo_vector(filename):
    fig = plt.figure()
    gs = fig.add_gridspec(len(array_length_list), hspace=0.5)
    axs = gs.subplots()
    fig.set_size_inches(4, 20)

    with open(filename, "r") as file:
        line_num = 0
        line = list(map(float, file.readline().split(',')[:-1]))
        while len(line) > 1:
            axs[line_num].set(xlabel="размер блока", ylabel="время работы в секундах")
            axs[line_num].set_title(f"Размер векторов равен 2^{array_length_list[line_num]}")
            axs[line_num]
            axs[line_num].plot([2**i for i in array_block_sizes_list], line)

            line_num += 1
            line = list(map(float, file.readline().split(',')[:-1]))
```

```python
def foo_matrix_2d_block(filename):
    fig = plt.figure()
    gs = fig.add_gridspec(len(matrices_size_list), hspace=0.5)
    axs = gs.subplots()
    fig.set_size_inches(4, 20)

    with open(filename, "r") as file:
        line_num = 0
        line = list(map(float, file.readline().split(',')[:-1]))
        while len(line) > 1:
            axs[line_num].set(xlabel="размер блока", ylabel="время работы в секундах")
            axs[line_num].set_title(f"Число элементов в матрицах 2^{matrices_size_list[line_num] * 2}")
            axs[line_num].plot([f"({2**i}, {2**i})" for i in matrices_block_sizes_list], line)

            line_num += 1
            line = list(map(float, file.readline().split(',')[:-1]))
```
